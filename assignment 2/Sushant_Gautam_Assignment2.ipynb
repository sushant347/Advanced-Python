## Question 6: List Sanitization and Metrics

```python
# Question 6: List Sanitization and Metrics
def sanitize_marks(marks_list):
    """
    Remove invalid entries from marks list and return statistics
    Returns: (cleaned_list, invalid_count, valid_percentage)
    """
    cleaned = []
    invalid_count = 0

    for mark in marks_list:
        # Check if mark is valid (numeric and within range 0-100)
        if isinstance(mark, (int, float)) and not isinstance(mark, bool):
            if 0 <= mark <= 100:
                cleaned.append(mark)
            else:
                invalid_count += 1
        else:
            invalid_count += 1

    # Calculate percentage of valid marks
    total_entries = len(marks_list)
    valid_percentage = (len(cleaned) / total_entries * 100) if total_entries > 0 else 0

    return (cleaned, invalid_count, valid_percentage)

# Test the function
marks = [45, -3, 88, "NA", 102, None, 67]
result = sanitize_marks(marks)

print(f"Original list: {marks}")
print(f"Cleaned list: {result[0]}")
print(f"Invalid entries count: {result[1]}")
print(f"Valid marks percentage: {result[2]:.2f}%")
```

## Question 7: Dictionary Usage

```python
# Question 7: Dictionary Usage
# Initialize empty dictionary
students = {}

def add_student(roll, name):
    """Add a new student to the dictionary"""
    students[roll] = name
    print(f"Added: {roll} -> {name}")

def update_student(roll, name):
    """Update existing student's name"""
    if roll in students:
        students[roll] = name
        print(f"Updated: {roll} -> {name}")
    else:
        print(f"Roll number {roll} not found!")

def display_students():
    """Display all students"""
    if students:
        print("\nStudent Records:")
        for roll, name in students.items():
            print(f"Roll: {roll}, Name: {name}")
    else:
        print("No students in the database!")

# Test the functions
add_student(101, "Amit")
add_student(102, "Sita")
add_student(103, "Hari")
display_students()

update_student(102, "Sita Sharma")
display_students()
```

## Question 8: Prime Number Function

```python
# Question 8: Prime Number Function
def is_prime(n):
    """
    Check if a number is prime
    Returns: True if prime, False otherwise
    """
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False

    # Check odd divisors up to square root of n
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

# Display prime numbers between 1 and 50
print("Prime numbers between 1 and 50:")
primes = [num for num in range(1, 51) if is_prime(num)]
print(primes)
```

## Question 9: Function Composition

```python
# Question 9: Function Composition
def calculate_total(marks):
    """Calculate total marks"""
    return sum(marks)

def calculate_average(marks):
    """Calculate average marks"""
    return sum(marks) / len(marks) if marks else 0

def classify_result(average):
    """Classify result based on average"""
    if average >= 75:
        return "Distinction"
    elif average >= 40:
        return "Pass"
    else:
        return "Fail"

def process_student_result(name, marks):
    """
    Combine all functions to process student result
    Returns: (name, total, average, result)
    """
    total = calculate_total(marks)
    average = calculate_average(marks)
    result = classify_result(average)

    return (name, total, average, result)

# Test the function
student_name = "Ram Kumar"
student_marks = [78, 82, 91, 85, 88]

result = process_student_result(student_name, student_marks)
print(f"Name: {result[0]}")
print(f"Total: {result[1]}")
print(f"Average: {result[2]:.2f}")
print(f"Result: {result[3]}")

print("\n--- Benefits of Modular Design ---")
print("1. Reusability: Each function can be used independently")
print("2. Maintainability: Easy to update individual functions")
print("3. Testing: Each function can be tested separately")
print("4. Readability: Code is organized and easy to understand")
print("5. Debugging: Easier to isolate and fix issues")
```

## Question 10: Defensive Input Parsing

```python
# Question 10: Defensive Input Parsing
def parse_student_data(input_string):
    """
    Parse student data from format: "Name|Roll|Age|Mark1, Mark2, Mark3"
    Returns: Dictionary with roll number as key
    """
    try:
        # Split by pipe character
        parts = input_string.split("|")

        if len(parts) != 4:
            return {"error": "Invalid format - expected 4 parts separated by |"}

        name = parts[0].strip()
        roll = parts[1].strip()
        age_str = parts[2].strip()
        marks_str = parts[3].strip()

        # Validate and convert age
        try:
            age = int(age_str)
            if age < 16:
                return {"error": f"Invalid age: {age}. Must be >= 16"}
        except ValueError:
            return {"error": f"Invalid age format: {age_str}"}

        # Parse and validate marks
        try:
            marks = [int(m.strip()) for m in marks_str.split(",")]

            # Validate each mark
            for mark in marks:
                if mark < 0 or mark > 100:
                    return {"error": f"Invalid mark: {mark}. Must be between 0-100"}
        except ValueError:
            return {"error": f"Invalid marks format: {marks_str}"}

        # Return valid data with roll number as key
        return {
            roll: {
                "name": name,
                "age": age,
                "marks": marks
            }
        }

    except Exception as e:
        return {"error": f"Unexpected error: {str(e)}"}

# Test with valid input
valid_input = "Ram|101|22|78, 88, 91"
result1 = parse_student_data(valid_input)
print("Valid input test:")
print(result1)

# Test with malformed inputs
print("\nMalformed input tests:")
print(parse_student_data("Ram|101|15|78"))  # Age < 16
print(parse_student_data("Ram|101|22|78, 105, 91"))  # Mark > 100
print(parse_student_data("Ram|101|abc|78, 88"))  # Invalid age
print(parse_student_data("Ram|101"))  # Missing parts
```

## Question 11: Type Stability and Conversion

```python
# Question 11: Type Stability and Conversion
def separate_and_convert(data):
    """
    Separate numeric and non-numeric values
    Convert numeric values to float
    Returns: (numeric_list, non_numeric_list)
    """
    numeric = []
    non_numeric = []

    for item in data:
        # Check if item is numeric (int, float, or bool)
        # Note: isinstance(True, int) returns True because bool is a subclass of int
        if isinstance(item, (int, float, bool)) and not isinstance(item, type(None)):
            numeric.append(float(item))
        else:
            non_numeric.append(item)

    return (numeric, non_numeric)

# Test the function
data = [10, "20", 30.5, "abc", None, True]
numeric_list, non_numeric_list = separate_and_convert(data)

print(f"Original data: {data}")
print(f"Numeric values (as float): {numeric_list}")
print(f"Non-numeric values: {non_numeric_list}")

print("\n--- Explanation ---")
print("True behaves as a numeric value because:")
print("1. bool is a subclass of int in Python")
print("2. True has a numeric value of 1, and False has a value of 0")
print("3. This allows boolean values to be used in arithmetic operations")
print(f"Example: True + 5 = {True + 5}, False + 5 = {False + 5}")
```

## Question 14: Dictionary Aggregation Logic

```python
# Question 14: Dictionary Aggregation Logic
def analyze_performance(records):
    """
    Compute average per student and identify top/bottom performers
    Returns: Dictionary with averages and performer info
    """
    result = {
        "averages": {},
        "top_performer": None,
        "bottom_performer": None
    }

    # Calculate averages
    for student, marks in records.items():
        avg = sum(marks) / len(marks)
        result["averages"][student] = round(avg, 2)

    # Find top and bottom performers
    if result["averages"]:
        top_student = max(result["averages"], key=result["averages"].get)
        bottom_student = min(result["averages"], key=result["averages"].get)

        result["top_performer"] = {
            "name": top_student,
            "average": result["averages"][top_student]
        }
        result["bottom_performer"] = {
            "name": bottom_student,
            "average": result["averages"][bottom_student]
        }

    return result

# Test the function
records = {
    "Amit": [78, 82, 91],
    "Sita": [88, 79, 85],
    "Hari": [45, 52, 60]
}

analysis = analyze_performance(records)

print("Performance Analysis:")
print(f"\nStudent Averages:")
for student, avg in analysis["averages"].items():
    print(f"  {student}: {avg}")

print(f"\nTop Performer: {analysis['top_performer']['name']} with average {analysis['top_performer']['average']}")
print(f"Bottom Performer: {analysis['bottom_performer']['name']} with average {analysis['bottom_performer']['average']}")
```

## Question 15: Nested Data Structure Analysis

```python
# Question 15: Nested Data Structure Analysis
def analyze_school_data(school_data):
    """
    Analyze nested dictionary of school grades
    Returns: Dictionary with class toppers, school average, and student averages

    Nested dictionaries are useful for hierarchical data because:
    1. They represent real-world organizational structures (school -> class -> student)
    2. Allow easy access to data at different levels
    3. Make relationships between entities explicit
    4. Enable efficient data grouping and retrieval
    """
    result = {
        "class_toppers": {},
        "school_average": 0.0,
        "student_averages": {}
    }

    all_marks = []

    # Process each class
    for class_name, students in school_data.items():
        class_best_avg = 0
        class_best_student = None

        # Process each student in the class
        for student_name, marks in students.items():
            # Calculate student average
            avg = sum(marks) / len(marks)
            result["student_averages"][student_name] = round(avg, 2)

            # Track all marks for school average
            all_marks.extend(marks)

            # Track class topper
            if avg > class_best_avg:
                class_best_avg = avg
                class_best_student = student_name

        # Store class topper
        result["class_toppers"][class_name] = (class_best_student, round(class_best_avg, 2))

    # Calculate school average
    if all_marks:
        result["school_average"] = round(sum(all_marks) / len(all_marks), 2)

    return result

# Test the function
school_data = {
    "Class A": {
        "Amit": [78, 82, 91],
        "Sita": [88, 79, 85]
    },
    "Class B": {
        "Hari": [45, 52, 60],
        "Ram": [92, 88, 95]
    }
}

analysis = analyze_school_data(school_data)

print("School Analysis Report")


print("\nStudent Averages:")
for student, avg in analysis["student_averages"].items():
    print(f"  {student}: {avg}")

print("\nClass Toppers:")
for class_name, (student, avg) in analysis["class_toppers"].items():
    print(f"  {class_name}: {student} ({avg})")

print(f"\nOverall School Average: {analysis['school_average']}")

print("\n--- Why Nested Dictionaries Are Useful ---")
print("1. Hierarchical Organization: Mirrors real structure (School > Class > Student)")
print("2. Easy Navigation: Access any level using keys")
print("3. Flexible: Can add/remove classes or students easily")
print("4. Maintainable: Clear data relationships")
print("5. Scalable: Can handle large organizational structures")
```
